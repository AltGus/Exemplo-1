```python
# Passo 2: Importar o Dataset e Printa-lo (Library Books)

import pandas as pd

# Caminho do dataset (ajuste conforme o local no seu Colab)
dataset_path = "/content/sample_data/Books_data - Sheet1.csv"

# Lendo o arquivo CSV
df = pd.read_csv(dataset_path)

# Exibindo as primeiras linhas do dataset
print("Visualizando as primeiras linhas do dataset:")
display(df.head())

# Exibindo informações gerais do dataset
print("\nInformações gerais do dataset:")
print(df.info())

# Exibindo estatísticas descritivas básicas
print("\nResumo estatístico dos dados numéricos:")
display(df.describe())
```

# 3 - Filtrar Colunas Interessantes

```python
# Passo 3: Formatação e Limpeza dos Dados (Library Books)

# Verificando se há valores nulos em cada coluna
print(" Verificando valores nulos no dataset:\n")
print(df.isnull().sum())

# Substituindo valores nulos em colunas de texto por 'Desconhecido' (se existirem)
df = df.fillna('Desconhecido')

# Exemplo: convertendo colunas de data, se existirem
# (ajuste os nomes das colunas conforme o seu dataset)
colunas_datas = ['Date of Publication', 'Year', 'Issued Date', 'Return Date']

for coluna in colunas_datas:
    if coluna in df.columns:
        df[coluna] = pd.to_datetime(df[coluna], errors='coerce')

# Conferindo os tipos de dados após a formatação
print("\n Tipos de dados após formatação:")
print(df.dtypes)

# Verificando novamente se ainda restam valores nulos
print("\n Valores nulos após tratamento:")
print(df.isnull().sum())

# Visualizando uma amostra dos dados após limpeza
print("\n Amostra dos dados limpos:")
display(df.head())
```

# 4 - Train Test Split

```python
# Passo 4: Implementação dos Algoritmos de Recomendação


import random
# Algoritmo Aleatório

# Seleciona os livros que estão disponíveis (status = 'available' ou similar)
livros_disponiveis = df[df['status'].str.lower().isin(['available', 'returned', 'devolvido'])]['title'].unique()

def recomendar_livro_aleatorio(livros_disponiveis):
    """
    Retorna um livro aleatório disponível.
    """
    if len(livros_disponiveis) == 0:
        print(" Nenhum livro disponível no momento.")
        return None
    livro_recomendado = random.choice(livros_disponiveis)
    print(f" Livro recomendado aleatoriamente: {livro_recomendado}")

# Testando o algoritmo aleatório
recomendar_livro_aleatorio(livros_disponiveis)


# Contando quantas vezes cada livro aparece (mais emprestado = mais popular)
livros_populares = df['title'].value_counts()

def recomendar_livro_popular(livros_populares):
    """
    Retorna o livro mais popular (mais emprestado).
    """
    if livros_populares.empty:
        print(" Nenhum registro de empréstimo encontrado.")
        return None
    livro_popular = livros_populares.idxmax()
    emprestimos = livros_populares.max()
    print(f" Livro recomendado heuristicamente (mais popular): {livro_popular}")
    print(f" Total de empréstimos: {emprestimos}")

# Testando o algoritmo heurístico
recomendar_livro_popular(livros_populares)
```

# 5 - Modelo para Treinar

```python
#  Passo 5: Implementação com Aprendizado de Máquina (ML) — Versão em Português

import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Baixar a lista de stopwords em português (só precisa rodar uma vez)
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_pt = stopwords.words('portuguese')

# Criando a coluna com informações combinadas
df['info'] = df['author'].astype(str) + " " + df['category'].astype(str)

# Vetorizando os textos com stopwords em português
vectorizer = TfidfVectorizer(stop_words=stopwords_pt)
tfidf_matrix = vectorizer.fit_transform(df['info'])

# Calculando similaridade
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Função para recomendar livros semelhantes
def recomendar_livro_similar(titulo, df=df, cosine_sim=cosine_sim):
    if titulo not in df['title'].values:
        print(f" O livro '{titulo}' não foi encontrado no dataset.")
        return

    idx = df.index[df['title'] == titulo][0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    similar_indices = [i[0] for i in sim_scores]

    print(f" Livros semelhantes a '{titulo}':")
    display(df.iloc[similar_indices][['title', 'author', 'category']])

# Teste com título (pode trocar para o nome de um livro do seu dataset)
recomendar_livro_similar("Steve Jobs")
```

# 6 - Modelo para Avaliar




```python
#  Passo 6: Implementação com Deep Learning (Rede Neural)


import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


# Criar uma coluna combinada com autor + categoria
df['info_text'] = df['author'].astype(str) + " " + df['category'].astype(str)

# Converter o status ("issued" / "available") para números
label_encoder = LabelEncoder()
df['status_encoded'] = label_encoder.fit_transform(df['status'])

# Separar textos (X) e rótulos (y)
X = df['info_text'].values
y = df['status_encoded'].values

# Tokenização — transformar texto em números
tokenizer = Tokenizer(num_words=1000, oov_token="<OOV>")
tokenizer.fit_on_texts(X)
sequences = tokenizer.texts_to_sequences(X)
padded = pad_sequences(sequences, maxlen=20, padding='post')

# Dividir treino e teste
X_train, X_test, y_train, y_test = train_test_split(padded, y, test_size=0.2, random_state=42)


model = Sequential([
    Embedding(input_dim=1000, output_dim=16, input_length=20),
    GlobalAveragePooling1D(),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')  # Saída binária
])

# Compilar modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


# Treinamento do modelo

print(" Treinando a rede neural...")
history = model.fit(X_train, y_train, epochs=15, validation_data=(X_test, y_test), verbose=1)

# Avaliação do modelo

loss, acc = model.evaluate(X_test, y_test)
print(f"\n Acurácia do modelo de Deep Learning: {acc*100:.2f}%")

# Teste de previsão

# Exemplo de teste — prever status de um novo livro
novo_texto = ["Machado de Assis Romance"]
seq = tokenizer.texts_to_sequences(novo_texto)
pad = pad_sequences(seq, maxlen=20, padding='post')

pred = model.predict(pad)
classe = "Disponível" if pred[0][0] < 0.5 else "Emprestado"

print(f"\n Previsão para o livro '{novo_texto[0]}': {classe}")
```

# 7 — Visualizar o desempenho do modelo

```python
# Passo 7: Visualização do Desempenho do Modelo

import matplotlib.pyplot as plt

# Gráfico da perda (loss)
plt.figure(figsize=(10,4))
plt.plot(history.history['loss'], label='Treino', marker='o')
plt.plot(history.history['val_loss'], label='Validação', marker='o')
plt.title('Evolução da Função de Perda (Loss)')
plt.xlabel('Épocas')
plt.ylabel('Perda')
plt.legend()
plt.grid(True)
plt.show()

# Gráfico da acurácia
plt.figure(figsize=(10,4))
plt.plot(history.history['accuracy'], label='Treino', marker='o')
plt.plot(history.history['val_accuracy'], label='Validação', marker='o')
plt.title('Evolução da Acurácia do Modelo')
plt.xlabel('Épocas')
plt.ylabel('Acurácia')
plt.legend()
plt.grid(True)
plt.show()
```

# 8 — Salvando e Carregando o Modelo de Deep Learning

```python
# Passo 8: Salvando e Carregando o Modelo Treinado

from tensorflow.keras.models import load_model


# 1- Salvando o modelo


# Salvar o modelo treinado em um arquivo .h5
model.save("modelo_recomendacao_livros.h5")
print(" Modelo salvo com sucesso em 'modelo_recomendacao_livros.h5'!")

# 2- Carregando o modelo salvo


# Carregar o modelo salvo
modelo_carregado = load_model("modelo_recomendacao_livros.h5")
print(" Modelo carregado novamente com sucesso!")

# 3- Testando o modelo carregado

# Testar o modelo com um novo exemplo (mesmo formato do Passo 6)
novo_texto_teste = ["Machado de Assis Romance"]
seq_teste = tokenizer.texts_to_sequences(novo_texto_teste)
pad_teste = pad_sequences(seq_teste, maxlen=20, padding='post')

# Fazer previsão
pred_teste = modelo_carregado.predict(pad_teste)
classe_prevista = "Disponível" if pred_teste[0][0] < 0.5 else "Emprestado"

print(f"\n Previsão usando modelo carregado: {classe_prevista}")
```
